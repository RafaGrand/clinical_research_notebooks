```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import matplotlib.pyplot as plt
import pandas as pd
import sidetable
pd.set_option('display.max_rows',None)

file_path = 'https://github.com/RafaGrand/clinical_research_notebooks/blob/master/censo.xlsx'
df = pd.read_excel(file_path, header=0, usecols=[0,4,8,10,11,12], names=['PABELLON','EPS','EDAD','DIAGNOSTICO','FECHA INGRESO','DIAS ESTANCIA'])
filter_df = df[df['DIAGNOSTICO'].str.contains("Coronavirus") | df['DIAGNOSTICO'].str.contains("COVID")]
filter_df.stb.freq(['DIAGNOSTICO'], style=True)


colors  = ("steelblue","salmon", "palevioletred", "dodgerblue", "seagreen", "plum", "blue", "indigo", "beige", "yellow")

i = 0 #counter

for col in filter_df: 
    sizes = filter_df[col].value_counts()
    pie = filter_df[col].value_counts().plot(kind='pie',colors=colors,shadow=True,autopct='%1.1f%%',startangle=30,radius=1.5,
                                      center=(0.5,0.5),textprops={'fontsize':12},frame=False,labels=None,pctdistance=.65)
    labels = sizes.index.unique()
    plt.gca().axis("equal")
    plt.title(df.columns[i],weight='bold',size=14)
    plt.subplots_adjust(left=0.0,bottom=0.1,right=0.85)
    #plt.savefig(str(df.columns[i])+'.png',dpi=100,bbox_inches='tight')
    pie.set_ylabel('')
    plt.legend(labels,bbox_to_anchor=(0.5, -.2),
           bbox_transform=plt.gcf().transFigure)
    i = i+1
    plt.show()
```
